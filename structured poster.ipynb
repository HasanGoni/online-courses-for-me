```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

# Function decleration

```python
def some_thing(gg):
    gg=gg.replace(',','.')
    gg = gg.replace(';',',')
    gg = gg.split(' ')
    listo=[ ]
    for i in gg:
        hh=i.replace(',',' ') 
        hh=float(hh)
        listo.append(hh)
    return listo
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(tick_number)
    if N == 0:
        return r"$Day 0$"
    elif N == 1:
        return r"$Day 1$"
    elif N == 2:
        return r"$Day 2$"
    elif N == 3:
        return r"$Day 3$".format(N)
    elif N == 5:
        return r"$Day 5$".format(N)
    elif N == 7:
        return r"$Day 7$".format(N)
    else:
        return " "
    
     
```

# LAB


```python
##raw excel value
back1_25 = '1,27; 1,67; 1,74; 1,98; 1,63; 1,66'
back1_30 = '1,20; 1,50; 1,74; 2,17; 1,39; 1,28'
back1_35 = '1,16; 1,50; 1,68; 2,09; 1,55; 1,45'
##comma separated value
back1_25 = some_thing(back1_25)
back1_30 = some_thing(back1_30)
back1_35 = some_thing(back1_35)
```

```python
def from_list_to_figure(back1_25, back1_30, back1_35, desired_name):    
    day_num = ['day0', 'day1', 'day2', 'day3', 'day5', 'day7']
    t1 = pd.Series( [pd.Timestamp('2018-09-01'), pd.Timestamp('2018-09-02'), pd.Timestamp('2018-09-03'), pd.Timestamp('2018-09-04'), pd.Timestamp('2018-09-06'), pd.Timestamp('2018-09-08')])
    t1 = pd.DataFrame(t1)
    t1.loc[:,'back1_25'] =back1_25
    t1.loc[:,'back1_30'] = back1_30
    t1.loc[:, 'back1_35'] = back1_35
    t1.rename(columns = {0:'Date'}, inplace = True)
    t1.set_index('Date', inplace = True)
    plt.figure(figsize=(10,8))
    plt.style.use('ggplot')
    ax=plt.axes()
    clrs = sns.color_palette('husl', n_colors=len(t1.columns)) 
    j=0
    for i in t1.columns:
            lines = ax.plot(t1.loc[:,i], 'o-')

            lines[0].set_color(clrs[j])
            j+=1
    
    plt.rc('xtick', labelsize=12) 
    plt.rc('ytick', labelsize=12) 
    plt.rc('legend', fontsize=12)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
                  fancybox=True, shadow=True, ncol=5)
    plt.xlim(736938,736945)
    ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
    plt.xlabel('Time')
        #******************************************************************************************************
    if desired_name == 'LAB':
        plt.ylabel('$ cfu/g (10^8)$')
    elif desired_name == 'Yeast':
         plt.ylabel('$ cfu/g (10^7)$')
    elif desired_name == 'Theamine':
        plt.ylabel('$μg/100g$')
    elif desired_name == 'Riboflavin':
         plt.ylabel('$μg/100g$')
    elif desired_name == 'pH':
         plt.ylabel('$pH$')
    plt.subplots_adjust (bottom = 0.2)
    
    if not os.path.exists('lopa_poster_structure'):
        os.mkdir('lopa_poster_structure')
    plt.savefig('lopa_poster_structure/'+desired_name+'.png')
```

```python
from_list_to_figure(back1_25, back1_30, back1_35, desired_name = 'LAB')
```
